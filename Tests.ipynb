```python
matrix = [ 
  'AUDACIAAPENF',
  'GZYYQHWNUMYX',
  'NADSLCADXQCF',
  'ABFIEJFRASES',
  'CEDEMRGOMXVO',
  'TEFJSWOILETE',
  'COMBOIODBHCR',
  'OBSTANTEAZAE',
  'XDXWQXNWXODS',
  'SDIAMETROAVC',
  'OIYVSJJDDRLR',
  'TSNSAQAILGZE',
  'RTEALCUNHATV',
  'ERFNIQENZMXE',
  'BAHDISNRQAJR',
  'OLANYTWRPSKQ',
  'CRIMINALISTA',
  'DIDWDTBVLAVZ']
word1 = 'CRIMINALISTA'
word2 = 'COMBO'
word3 = 'SOBRESCREVER'
```

```python
lword1 = len(word1)
lword2 = len(word2)



def get_possible_words(matrix, word):
    lword = len(word)
    nrows = len(matrix)
    ncols = len(matrix[0])
    
    # dir left to right
    words_left_to_right = []
    for i in range(nrows):
        for j in range(ncols - lword + 1):
            words_left_to_right.append(((i, j), matrix[i][j:j+lword]))

    # Transpose Matrix
    matrix_trans = list(zip(*matrix))
    for i, row in enumerate(matrix_trans):
        matrix_trans[i] = ''.join(row)
    nrows = len(matrix_trans)
    ncols = len(matrix_trans[0])

    # dir up to down
    words_up_to_down = []
    for i in range(nrows):
        for j in range(ncols - lword + 1):
            words_up_to_down.append(((i, j), matrix_trans[i][j:j+lword]))

    return words_left_to_right, words_up_to_down
```
